# 1. Giới thiệu về bài toán dự báo chuỗi thời gian
- Chuỗi thời gian là một tập các điểm dữ liệu phụ thuộc thời gian.
- Tập dữ liệu lí tưởng là tập các dữ liệu thời gian với các khoảng thời gian cách đều nhau (mỗi phút hoặc mỗi giây).
- Dự báo chuỗi thời gian (Time series forecasting): Phân tích mẫu dữ liệu quá khức để dự đoán tương lai.
- Một vài ví dụ:
+ Dự báo giá chứng khoán
+ Mùa bán hàng quần áo
+ Dự báo thời tiết
- Một số đặc trưng bài toán: Xu hướng (Trend) và mùa vụ (Seáonality).
